```python
import requests

url = "https://api.coingecko.com/api/v3/coins/markets?vs_currency=usd&per_page=20&price_change_percentage=1h%2C24h%2C7d%2C30d%2C200d%2C1y"

headers = {
    "accept": "application/json",
    "x-cg-demo-api-key": "CG-mNvSZ73GNiENGVUktW3QXEPp"
}

response = requests.get(url, headers=headers)
    
data = response.json()
```

```python
def api_runner():
    import requests
    global df
    url = "https://api.coingecko.com/api/v3/coins/markets?vs_currency=usd&per_page=20&price_change_percentage=1h%2C24h%2C7d%2C30d%2C200d%2C1y"
    headers = {
        "accept": "application/json",
        "x-cg-demo-api-key": "CG-mNvSZ73GNiENGVUktW3QXEPp"
    }
    response = requests.get(url, headers=headers)
    data = response.json()

    df2 = pd.json_normalize(data)
    df2['timestamp'] = pd.to_datetime('now')
    df = pd.concat([df, df2])

    #df = pd.json_normalize(data)
    #df['timestamp'] = pd.to_datetime('now')
    #df

    #if not os.path.isfile(r"C:\Users\rafsa\OneDrive\Desktop\Pandas Alex\Crypto API.csv"):
    #    df.to_csv(r"C:\Users\rafsa\OneDrive\Desktop\Pandas Alex\Crypto API.csv", header = 'column_names')
    #else:
    #    df.to_csv(r"C:\Users\rafsa\OneDrive\Desktop\Pandas Alex\Crypto API.csv", mode = 'a', header = False) # mode a = append
```
